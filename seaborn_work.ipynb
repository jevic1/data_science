```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x='tip',y='total_bill',data=tips,hue='day',size='size',palette='YlGnBu')
```

```python
sns.histplot(tips.tip,kde=True,bins=10)
```

```python
sns.distplot(tips.tip)
```

```python
sns.barplot(x="sex",y="tip",data=tips)
```

```python
sns.boxplot(x="day",y="tip",data=tips,hue="sex",palette="YlGnBu")
```

```python
sns.pairplot(tips)
```

```python
tips.corr()
```

```python
#sns.heatmap(tips.corr(),annot=True,cmap='coolwarm')
sns.heatmap(tips.corr(),annot=True,cmap='icefire')
```

```python
sns.countplot(x='size',data=tips,hue='sex',palette='Set2')
```

```python
sns.countplot(y='size',data=tips)
```
