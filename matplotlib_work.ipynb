```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas=pd.read_csv('gas_prices.csv')
```

```python
plt.figure(figsize=(8,5))
#plt.plot(gas.Year, gas.USA,'r.-',label='USA')
#plt.plot(gas.Year, gas.Canada,'b.-',label='Canada')
#plt.plot(gas['Year'], gas['South Korea'],'g.-',label='South Korea')
countries_to_look_at=['Australia','Canada','South Korea','USA']
for country in gas:
    if country in countries_to_look_at:
        plt.plot(gas.Year,gas[country],label=country,marker='.')

plt.xticks(gas.Year[::3])

plt.legend()
plt.show()
```

```python
fifa=pd.read_csv('fifa_data.csv')
```

```python
fifa.head()
```

```python
bins=np.arange(40,110,10)

plt.hist(fifa.Overall)
plt.xticks(bins)
plt.show()
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
labels=['left','right']
plt.style.use("ggplot")
explode=[.1,.1]

plt.pie([left,right],labels=labels,explode=explode, autopct="%.2f %%",pctdistance=0.8)
plt.show()
```

```python
plt.style.use('ggplot')
barcelona = fifa.loc[fifa.Club=="FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa.Club=="Real Madrid"]["Overall"]
labels=["Barcelona",["Madrid"]]

plt.boxplot([barcelona,madrid],labels=labels)
plt.show()
```
